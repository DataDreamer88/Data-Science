## Firstly Using Simple Python Approach to filter and play with data

Importing CSV file

```python
import csv

with open("movies.csv") as m:
    data = list(csv.reader(m))
    print(data)
```

My First row is Header and 2nd row onwards is data

```python
#printing data
data[1:]
```

```python
#differentiating Header and Data

header = data[0]
data = data[1:]
```

```python
print(header)
print()
print(data)
```

```python
def calculate_rating_stats(data, industry=None):    # default parameter for industry is None i.e for all industries
    ratings = [] # empty list
    for row in data:
        if row[3]!='NULL' and (not industry or row[1]==industry): #if rating is not null and (industry value is empty or Bollywood/hollywood/any specific)
            ratings.append(float(row[3])) # add it to empty list
    max_rating = max(ratings)
    min_rating = min(ratings)
    avg_rating = sum(ratings) / len(ratings)
    return max_rating, min_rating, avg_rating
```

```python
max_rating, min_rating, avg_rating = calculate_rating_stats(data) #no need to write other parameter for "all Industries" as it is defined as none by default
print(f"All records: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")
    
max_rating, min_rating, avg_rating = calculate_rating_stats(data, industry="Bollywood") #when industry is bollywood
print(f"Bollywood: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")
    
max_rating, min_rating, avg_rating = calculate_rating_stats(data, industry="Hollywood") #when industry is hollywood
print(f"Hollywood: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")
```
